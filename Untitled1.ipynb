```python
#with as

df = pd.read_sql('''
                  with tab as
                  (select model, speed, hd, ram, price,
                    case when speed = 600 then 1
                      else 0 
                        end flag from Laptop)
                  select * from tab 
                    where ram > 32
                        ''', conn_sqlite)
df
```

```python
#для in + like (desc - по убыванию asc - по возрастанию)

df = pd.read_sql('''with tab as
                    (select model, price, speed,
                    case when price > 400 
                    then 1 
                    else 0 
                    end flag from PC
                    where model like "12%"
                    order by price asc)
                    select * from tab 
                      where model like "123%"
                      ''', conn_sqlite)
df
```

```python
#Найдите номер модели продукта (ПК, ПК-блокнота или принтера), имеющего самую высокую цену. Вывести: model
df = pd.read_sql('''with tab as  
                    (select model, price
                          from Laptop 
                          where price = (select MAX(price)
                                        from Laptop)
                    union
                    select model, price
                          from PC 
                          where price = (select MAX(price)
                                        from PC)
                    union
                    select model, price
                          from Printer
                          where price = (select MAX(price)
                                        from Printer)
                    )
                    select model from tab
                      where price = (select MAX(price)
                      from tab)
                      ''', conn_sqlite)
df
```

```python
#Найдите номер модели продукта (ПК, ПК-блокнота или принтера), имеющего самую высокую цену. Вывести: model
df = pd.read_sql(''' select product.model, Laptop.price, PC.price,Printer.price from product 
                      left join Laptop on product.model = Laptop.model                    
                      left join PC on product.model = PC.model
                      left join Printer on product.model = Printer.model
                    
                      ''', conn_sqlite)
df
```
